```python
# Chargement des librairies
from matplotlib import pyplot
import matplotlib
import math
import sklearn.metrics as sm
from sklearn.cluster import KMeans
from sklearn import datasets
import numpy as np
from scipy import stats
import pandas as pd
import ipywidgets as widgets
from ipywidgets import interact
import datetime
from sklearn import datasets
from sklearn import decomposition
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
import random
```

```python
tfin = 1500
date = np.zeros(tfin, dtype = float)
hosp = np.zeros(tfin, dtype = float)
entree = np.zeros(tfin, dtype = float)
sortie = np.zeros(tfin, dtype = float)
kentree = np.zeros(tfin, dtype = float) # coeff d'augmentation des entrées (en lien avec nb reproduction)

# Paramètre
tcont = 7 # durée où l'on est contagieux
thosp = 12 # durée d'hospitalisation
tarrivee = 10 # durée pour arriver à l'hopital (retard dans la mesure)
ndeconf = 1.2 # nombre de reproduction en déconfinement
nconf = 0.7 # nombre de reproduction en confinement

# initialisation
etatconf = 'deconf'
entree[0] = 1000
hosp[0] = 1000
vk = 0.0 #variation aléatoire sur le nombre de reproduction
```

```python
for t in range(1,tfin,1):
    
    # regulation
    if t <=tarrivee : # initialisation
        kentree[t] = pow(ndeconf,1/tcont) 
        kentree[t+tarrivee] = pow(ndeconf,1/tcont) 
    elif t >= tfin-tarrivee : kentree[t] =  kentree[t] # fin
    elif  hosp[t-1] > 35000/pow(1.3,tarrivee/tcont):
        kentree[t+tarrivee] = pow(nconf,1/tcont) # confinement
    elif hosp[t-1] < 10000 : kentree[t+tarrivee] = pow(0.95,1/tcont)
    else:kentree[t+tarrivee] = pow(0.8,1/tcont)

    
    # perturbation : le nombre de reproduction varie tous les mois
    if int(t/30) == t/30:
        vk = random.choices([0, 0.1, 0.2,0.3], weights = [4,3, 2, 1], k = 1)
    kentree[t] = pow(pow(kentree[t],tcont) + vk,1/tcont)
     
    # evolution
    entree[t] = entree[t-1] * kentree[t]
    if t >= thosp : sortie[t] = entree[t-thosp]
    hosp[t] = hosp[t-1] + entree[t] - sortie[t]
    date[t] = t
    
```

```python
pyplot.figure(figsize=(15,5))
pyplot.ylim(0,80000)
pyplot.gca().xaxis.set_ticks(range(0,tfin,30))
pyplot.gca().yaxis.set_ticks(range(0,80000,5000))

pyplot.plot(date,hosp,color='black',label ='hospitalisation')
pyplot.plot(date,entree,'red',label ='entree')
pyplot.plot(date,sortie,'green',label = 'sortie')
pyplot.axhline(y = 50000,color = 'red')
pyplot.axhline(y = 35000,color = 'black')
pyplot.axhline(y = 35000/pow(max(ndeconf,1.1),tarrivee/tcont),color = 'green')
pyplot.grid()
pyplot.legend(loc ='upper left')

pyplot.gca().twinx()
pyplot.ylim(0,1.6)
pyplot.plot(date,pow(kentree,tcont),label ='nombre de reproduction')
pyplot.legend()

pyplot.show()
```

```python
tf = 1000
ng = 2 # nombre de groupe
popg = 100000 # population groupe
ti = np.zeros(tf, dtype = float)
sain = np.zeros((ng,tf), dtype = float)
malade = np.zeros((ng,tf), dtype = float)
new_malade = np.zeros((ng,tf), dtype = float)
gueri = np.zeros((ng,tf), dtype = float)

rint = 1.2 # R0 dans le groupe
tc = 7 # durée de contamination
rext = 0.05
# R0 entre les groupes

new_malade[0,0] = 50 # initialisation
new_malade[1,0] = 50 # initialisation
malade[:,0] = new_malade[:,0]
sain[:,0] = popg - new_malade[:,0]
seuil = 10

for t in range(1, tf):
    for g in range(ng):
        # infection dans le groupe
        if malade[g,t-1] >= seuil :new_malade[g,t] = (malade[g,t-1]-seuil) * rint/tc  * sain[g,t-1] / popg
        # infection hors du groupe
        for gext in range(ng):
            if gext != g:
                if malade[gext,t-1] >= seuil :new_malade[g,t] =new_malade[g,t] + (malade[gext,t-1]-seuil) * rext/tc  * sain[g,t-1] / popg
        
        if t > 7 :new_gueri  = new_malade[g, t-7]
        else:new_gueri  = new_malade[g, 0] / 7
        
        malade[g,t] = malade [g,t-1]  +  new_malade[g,t] - new_gueri
        sain[g,t] = sain[g,t-1] - new_malade[g,t]
        gueri[g,t] = gueri[g,t-1 ] + new_gueri
     
    ti[t] = t
```

```python
tmax = 200
pyplot.figure(figsize=(15,5))
pyplot.subplot(1,2,1)
pyplot.plot(ti[0:tmax ],sain[0,0:tmax ],'blue')
pyplot.plot(ti[0:tmax ],gueri[0,0:tmax ],'green')
pyplot.plot(ti[0:tmax ],gueri[1,0:tmax ],'green',linestyle ='--')
pyplot.ylim(0,100000)
pyplot.grid()
axe2 = pyplot.gca().twinx()
axe2.plot(ti[0:tmax ],malade[0,0:tmax ],'red')
axe2.plot(ti[0:tmax ],new_malade[0,0:tmax ],'red',linestyle ='--')


pyplot.subplot(1,2,2)
pyplot.plot(ti[0:tmax ],sain[1,0:tmax ],'blue')
pyplot.plot(ti[0:tmax ],gueri[1,0:tmax ],'green')
pyplot.ylim(0,100000)
pyplot.grid()
axe2 = pyplot.gca().twinx()
axe2.plot(ti[0:tmax ],malade[1,0:tmax ],'red')
axe2.plot(ti[0:tmax ],new_malade[1,0:tmax ],'red',linestyle ='--')
```
