```python
# fonction de comparaison de deux courses
def comparaison(ca,cb):
    print(),print();print('Comparaison de',ca.at[0,'trail'],'avec',cb.at[0,'trail']);print()
    im = 0;jm=0; nab = 0;Smina = 0 ; Sminb = 0
    for i in range(len(ca)):
        for j in range(len(cb)):
            if (ca.at[i,'nom'] == cb.at[j,'nom'] 
                and ca.at[i,'prenom'] == cb.at[j,'prenom'] 
                and ca.at[i,'nom'] != 'MARCOVICI'):
                ta = ca.at[i,'temps']
                tb = cb.at[j,'temps']
                mina= (int(str.split(ta,'h')[0])*60 + 
                        int(str.split(str.split(ta,'h')[1],"'")[0]) + 
                       int(str.split(str.split(ta,'h')[1],"'")[1])/60)
                minb= (int(str.split(tb,'h')[0])*60 + 
                       int(str.split(str.split(tb,'h')[1],"'")[0]) + 
                       int(str.split(str.split(tb,'h')[1],"'")[1])/60)
                moy = (mina+minb)/2
                mina = mina / moy
                minb = minb  / moy
                Smina = Smina + mina
                Sminb = Sminb + minb
                nab = nab  + 1
                print(ca.at[i,'nom'].ljust(12),ca.at[i,'prenom'].ljust(12),
                      ta.ljust(9),tb.ljust(9),''.ljust(9),round(minb/mina,2))
            elif (ca.at[i,'nom'] == cb.at[j,'nom'] 
                and ca.at[i,'nom'] == 'MARCOVICI'):im = i;jm = j
    if nab >= 1 : kab = Sminb / Smina
    else: kab = 0
    if im > 0 and jm > 0:
        ta = ca.at[im,'temps']
        mina= (int(str.split(ta,'h')[0])*60 + 
                        int(str.split(str.split(ta,'h')[1],"'")[0]) + 
                       int(str.split(str.split(ta,'h')[1],"'")[1])/60)
        if kab > 0 : minab = mina * kab
        else : minab = mina
        tab = repr(int(minab/60))+'h'+repr(int(minab - 60 * (int(minab/60)))).rjust(2)+"'"+repr(round((minab - int(minab))*60)).rjust(2)+"''"
        print('');print(ca.at[im,'nom'].ljust(12),ca.at[im,'prenom'].ljust(12),
          ca.at[im,'temps'],cb.at[jm,'temps'],tab.rjust(9),round(kab,2))
    else:
        print('');print(''.rjust(55),repr(round(kab,3)).ljust(4))
        
    return(kab,nab)
```
